```python
import pandas as pd
```

#### Reading Confirmed Cases dataset

```python
confirmed_cases_dataset = pd.read_csv("../../../data/stage_1/covid_confirmed_usafacts.csv")
confirmed_cases_dataframe = pd.DataFrame(confirmed_cases_dataset)
```

```python
confirmed_cases_dataframe
```

```python
confirmed_cases_dataframe.dtypes
```

#### Reading County population Data

```python
county_population_data = pd.read_csv("../../../data/stage_1/covid_county_population_usafacts.csv")
county_population_dataframe = pd.DataFrame(county_population_data)
```

```python
county_population_dataframe
```

```python
county_population_dataframe.dtypes
```

#### Reading Covid Deaths data

```python
death_facts = pd.read_csv("../../../data/stage_1/covid_deaths_usafacts.csv")
death_facts_dataframe = pd.DataFrame(death_facts)
```

```python
death_facts_dataframe
```

```python
death_facts_dataframe.dtypes
```

#### Merging Confirmed data and Deaths data

```python
merged_confirmed_deaths = pd.merge(confirmed_cases_dataframe, death_facts_dataframe, on=['countyFIPS','County Name','State','StateFIPS'])
merged_confirmed_deaths
```

#### Merging the above dataset with County Population

```python
merged_dataset = pd.merge(merged_confirmed_deaths,county_population_dataframe, on=['countyFIPS','State'])
merged_dataset
```

```python
merged = merged_dataset.drop(['County Name_y'],axis=1)
merged
```

```python
merged.to_csv('../../../data/stage_1/Covid_merged_dataset.csv',index = False)
```

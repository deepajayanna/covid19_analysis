```python
#Renaming 'County Name' to 'County Name_p' since there is a mismatch in county names in both the dataframes 
#and we want to do merge on countyFIPS and there is a memory issue when we are using on='countyFIPS'
population_df.rename(columns = {'County Name':'County Name_p'}, inplace = True)
#Converting all county names to lower Case and removing whitespaces(leading and trailing)
population_df['County Name_p'] = population_df['County Name_p'].apply(lambda cn: cn.lower().strip())
#Converting all State names to upper Case and removing whitespaces(leading and trailing)
population_df['State'] = population_df['State'].apply(lambda cn: cn.upper().strip())
population_df
```

#### Merge of Confirmed-death dataset(df) and population_df 

```python
covid_large_df= pd.merge(df,population_df)
covid_large_df
```

#### Removing the extra county name columns and displaying the final super covid dataframe

```python
covid_large_df=covid_large_df.drop(['County Name_d','County Name_p'], axis=1)
covid_large_df
```

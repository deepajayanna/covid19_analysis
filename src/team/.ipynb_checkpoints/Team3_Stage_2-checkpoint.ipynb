## Task 1: Compare Weekly Statistics

```python
import pandas as pd
```

```python
covid_cases = pd.read_csv('../../../data/stage_2/covid_confirmed_usafacts.csv')
covid_cases
```

### Remove all obersvations in which countyFIPS does not exist (i.e, countyFIPS == 0)

```python
ind = covid_cases[covid_cases['countyFIPS'] == 0].index
covid_cases.drop(ind, inplace=True)
```

### Remove observation dates in which most counties observed little to no COVID-19 cases

```python
# Drop observations where observed cases were scarce
covid_cases.drop(covid_cases.iloc[:, 4:57], inplace=True, axis=1)
```

### Create a copy dataframe with new indexes to preserve important variables ('countyFIPS', 'County Name', etc)

```python
new_ind = covid_cases.reset_index().set_index(['countyFIPS', 'County Name', 'State', 'StateFIPS'])
new_ind
```

```python
# Remove the old index
new_ind.drop(columns=['index'], inplace=True)
new_ind
```

### Sum all columns (date observations) to obtain number of total cases by days observed

```python
dates_sum = new_ind.iloc[:,].sum()
dates_sum
```

### Create a new dataframe containing a single row of the total number of cases by days observed

```python
total_dates_sum = pd.DataFrame(columns=covid_cases.columns)
total_dates_sum = total_dates_sum.append(dates_sum, ignore_index=True)
total_dates_sum
```

```python
# Remove irrelavent columns (i.e columns that don't contain covid case numbers)
total_dates_sum.drop(columns=['countyFIPS', 'County Name', 'State', 'StateFIPS'], inplace=True)
```

```python
total_dates_sum
```

### Get the new cases by day by taking the difference between the current day and the previous day

```python
new_cases_byday = total_dates_sum.diff(axis=1)
new_cases_byday
```

```python
# Remove that first observed date (since there is no observation before it, it will contain NaN value after subtraction)
new_cases_byday.drop(columns=['2020-03-15'], inplace=True)
new_cases_byday
```

```python
new_cases_byday.iloc[:, 21:28]
```

### Obtain the total number of weeks by dividing the number of observed days into groups of 7 (i.e, # of days in a week)

```python
num_weeks = len(new_cases_byday.columns)//7
```

### Sum every seven columns together to obtain total new cases per week (add results to a list)

```python
# Column counter variables (sum the total number of new cases for every 7 days)
i, n = 0, 7

# A list to add sums into
new_obs_byweek = []

# Loop from O to the number of total weeks (i.e, total number of observed days divided by 7)
for x in range(0, num_weeks): 
    # Sum columns i to n
    y = new_cases_byday.iloc[:, i:n].sum(axis=1)
    
    # Increment counter to point to next seven observations
    i+=7
    n+=7
    
    # Append results to list
    new_obs_byweek.append(y)
```

```python
byweek = pd.concat(new_obs_byweek, axis=1)
new_cases_byweek = byweek.iloc[:,].astype(int)
```

```python
new_cases_byweek
```

### Find mean number of new cases for each week observed

```python
i, n = 0, 7

# A list to add the means to
mean_obs_byweek = []
for x in range(0, num_weeks):
    y = round(new_cases_byday.iloc[:, i:n].mean(axis=1))
    i+=7
    n+=7
    mean_obs_byweek.append(y)
```

```python
mean_byweek = pd.concat(mean_obs_byweek, axis=1)
mean_cases_byweek = mean_byweek.iloc[:,].astype(int)
```

```python
mean_cases_byweek
```

### Find mean number of new cases for all weeks observed together

```python
mean_cases_allweeks = round(new_cases_byweek.iloc[:,].mean(axis=1)).astype(int)
mean_cases_allweeks
```

## Task 1.2: Compare data against other countries

#### Reading csv files

```python
from datetime import datetime
world_data = pd.read_csv('../../../data/stage_2/owid-covid-data.csv')
```

#### Checking Data

```python
world_data.head()
```

#### Selecting 5 countries with similar population to the United States

```python
world_data = world_data[world_data["location"].isin(["Indonesia", "Pakistan", "Nigeria", "Brazil", "Bangladesh"])]
world_data.head()
```

#### Sets the date of the data to a specific date range

```python
world_data = world_data[(world_data['date'] > '2020-03-21') & (world_data['date'] <= '2021-02-28')]
world_data.head()
```

#### Obtaining the Weekly data of the 5 countries

```python
#Gets the sum of cases and deaths for each week
world_data['date'] = world_data.date.apply(lambda d: datetime.strptime(d, '%Y-%m-%d'))
world_data_week_sum = world_data.groupby(["location", pd.Grouper(key='date', freq='w')]).sum().reset_index().round(decimals=0)
world_data_week_sum.head()
```

#### Selects data that will be used

```python
world_data_week_sum[["location", "date", "new_cases", "new_deaths", "population"]]
```

#### Uses United States Weekly cases from Task 1.1

```python
new_cases_byweek
```

#### Converts the data from wide to long format

```python
new_cases_byweek = pd.melt(new_cases_byweek, var_name='date', value_name='New_Cases')
new_cases_byweek.head()
```

#### Converts the date column values to the weekly dates

```python
new_cases_byweek['date'] = pd.to_datetime(new_cases_byweek['date'])

dates = pd.date_range(start='2020-03-22', end='2021-02-28', freq='W').to_frame()

dates.columns = ['date']

dates['date'] = pd.to_datetime(dates['date'], errors='coerce')

new_cases_byweek['date'] = dates['date'].values

new_cases_byweek.head()
```

#### Adding New Column location with value United States

```python
new_cases_byweek['location']='United States'
new_cases_byweek.head()
```

#### The total mean of cases for all the weeks

```python
#new_cases_byweek = new_cases_byweek.groupby("location")["New_Cases"].agg(["mean"]).reset_index().round(decimals=0)
#new_cases_byweek
new_cases_total_mean = new_cases_byweek.groupby("location")["New_Cases"].agg(["mean"]).reset_index().round(decimals=0)
new_cases_total_mean
```

#### The total mean of cases for each country

```python
world_data_week_cases_mean = world_data_week_sum.groupby("location")["new_cases"].agg(["mean"]).reset_index().round(decimals=0)
world_data_week_cases_mean
```

#### Merging the US cases mean with the other Countries

```python
world_cases_mean_final = pd.merge(world_data_week_cases_mean, new_cases_total_mean, how="outer")
world_cases_mean_final
```

As we can see from the table above The United States has the largest mean of cases followed by Brazil, Indonesia, Pakistan, Bangladesh, and Nigeria.

##### Calculating mean of new cases and new deaths of a week starting from week 2020-03-22 to 2021-02-28 for 5 world countries whose population is similar to United States of America.

```python
world_data_weekly_stats = world_data_week_sum[["location", "date", "new_cases", "new_deaths"]] 
world_data_week_newcases_mean = round(world_data_week_sum['new_cases'].div(7))
world_data_week_newdeaths_mean = round(world_data_week_sum['new_deaths'].div(7))
world_data_week_cases_temp_mean = world_data_week_newcases_mean.to_frame().rename(columns = {"new_cases": "new_cases_mean"})
world_data_week_deaths_temp_mean = world_data_week_newdeaths_mean.to_frame().rename(columns = {"new_deaths": "new_deaths_mean"})
print(world_data_week_newdeaths_mean)
print(world_data_week_cases_temp_mean)
print(world_data_week_deaths_temp_mean)
```

##### Concatenating weekly new cases and deaths mean of 5 countries to world data weekly sum data frame.

```python
world_data_week_newcases_mean = pd.concat([world_data_weekly_stats,world_data_week_cases_temp_mean["new_cases_mean"]], axis=1)
world_data_week_newcases_mean = pd.concat([world_data_week_newcases_mean,world_data_week_deaths_temp_mean["new_deaths_mean"]], axis=1)
print(world_data_week_newcases_mean)
```

##### Calculating mean of new cases and new deaths of a week starting from week 2020-03-22 to 2021-02-28 for United States of America.

```python
usa_week_newcases_mean = round(new_cases_byweek['New_Cases'].div(7))
usa_week_newdeaths_mean = round(deaths_obs_byweek['New_Deaths'].div(7))
usa_week_cases_temp_mean = usa_week_newcases_mean.to_frame().rename(columns = {"New_Cases": "new_cases_mean"})
usa_week_deaths_temp_mean = usa_week_newdeaths_mean.to_frame().rename(columns = {"New_Deaths": "new_deaths_mean"})
print(usa_week_cases_temp_mean)
usa_week_deaths_temp_mean
```

##### Concatenating weekly new cases and deaths mean of Unites States to new usa cases and deaths by week mean data frame to merge Unites States new cases and deaths by week mean information to 5 other world countries new cases and deaths by week mean data frame for comparision.

```python
usa_new_cases_deaths_byweek_mean = pd.concat([new_cases_byweek,usa_week_cases_temp_mean["new_cases_mean"]], axis=1)
usa_new_cases_deaths_byweek_mean = pd.concat([usa_new_cases_deaths_byweek_mean,deaths_obs_byweek["New_Deaths"]], axis=1) 
usa_new_cases_deaths_byweek_mean = pd.concat([usa_new_cases_deaths_byweek_mean,usa_week_deaths_temp_mean["new_deaths_mean"]], axis=1) 
usa_new_cases_deaths_byweek_mean = usa_new_cases_deaths_byweek_mean.rename(columns = {"New_Cases": "new_cases","New_Deaths": "new_deaths"})
cols_order = ['location', 'date', 'new_cases', 'new_deaths', 'new_cases_mean', 'new_deaths_mean']
usa_new_cases_deaths_byweek_mean = usa_new_cases_deaths_byweek_mean.reindex(columns=cols_order)
print(usa_new_cases_deaths_byweek_mean)
```

##### - Nigeria: ease of COVID-19 restrictions in the wake of economic fallout to open the businesses resulted peak of COVID-19 infections during                               3rd week of Jan 2021
##### - Pakistan: lack of practicing social distance by public resulted peak of COVID-19 cases during 3rd week of June 2020

## Number of cases Analysis  - Read the dataset

```python
import pandas as pd
covid_19_cases = pd.read_csv('../../../data/stage_2/covid_confirmed_usafacts.csv')
covid_19_cases
```

```python
# We have statewide unallocated for every State and countyFIPS is filled with 0 and it repeats for every Country. 
# let's clean up the dataset by removing those rows since we cannot access them uniquely for a state
dummyvalues = covid_19_cases.index[covid_19_cases['countyFIPS'] == 0].tolist()
covid_19_cases = covid_19_cases.drop(labels=dummyvalues,axis=0)
```

```python
covid_19_cases
```

```python
#Resetting the index to below columns so that they don't get summed up
refined_data = covid_19_cases.set_index(['countyFIPS','County Name','State','StateFIPS'])
refined_data
```

## Total number of cases per day

```python
# Generate total number of covid cases per day in the United States
covid_cases_summed = refined_data.sum(axis = 1, skipna = True)
covid_cases_summed
```

```python
new_covid_dataframe = pd.DataFrame(columns=refined_data.columns)
new_covid_dataframe = new_covid_dataframe.append(covid_cases_summed, ignore_index=True)
```

## Number of cases per day after calculating difference from previous day with current day

```python
newcases_perday = new_covid_dataframe.diff(axis=1)
newcases_perday
```

```python
newcases_perday.drop(columns=['2020-01-22'], inplace=True)
newcases_perday
```

## Calculate the total number of weeks for the dataset

```python
import math
number_of_weeks = math.floor(len(covid_cases_summed)/7)
number_of_weeks
```

## Total number of cases every week

```python
# This function calculates the total number of cases and average cases occured every week based on the operation value recieved
def calculate_sum_avg(dataset,operation):
    weekly_data = []
    i=0
    counter=7
    x=0
    while (x<number_of_weeks):
        if(operation == "sum"):
            batch = dataset.iloc[:,i:counter].sum(axis=1).astype(int)
        elif (operation == "avg"):
            batch = round(dataset.iloc[:,i:counter].mean(axis=1)).astype(int)
        i = i+7
        counter = counter+7
        x = x+1
        weekly_data.append(batch)
    return weekly_data
```

```python
IS_SUM = "sum"
weekly_summed_data = calculate_sum_avg(newcases_perday,IS_SUM)
total_cases = pd.concat(weekly_summed_data, axis=1)
total_cases
```

##  Average number of cases every week

```python
IS_MEAN = "avg"
weekly_averaged_data = calculate_sum_avg(newcases_perday,IS_MEAN)
new_avg = pd.concat(weekly_averaged_data, axis=1)
new_avg
```

##  calculating (mean, median, mode) for number of covid cases of every week taken together.

```python
mean_all_week = round(new_avg.mean(axis=1)).astype(int)
mean_all_week
```

```python
median_all_week = round(new_avg.median(axis=1)).astype(int)
median_all_week
```

```python
mode_all_week = round(new_avg.mode(axis=1)).astype(int)
mode_all_week
```

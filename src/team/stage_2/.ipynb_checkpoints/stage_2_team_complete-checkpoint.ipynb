## Number of cases Analysis  - Read the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt 
covid_19_cases = pd.read_csv('../../../data/stage_2/covid_confirmed_usafacts.csv')
covid_19_cases
```

```python
# We have statewide unallocated for every State and countyFIPS is filled with 0 and it repeats for every Country. 
# let's clean up the dataset by removing those rows since we cannot access them uniquely for a state
dummyvalues = covid_19_cases.index[covid_19_cases['countyFIPS'] == 0].tolist()
covid_19_cases = covid_19_cases.drop(labels=dummyvalues,axis=0)
```

```python
covid_19_cases
```

```python
#Resetting the index to below columns so that they don't get summed up
refined_data = covid_19_cases.set_index(['countyFIPS','County Name','State','StateFIPS'])
refined_data
```

## Total number of cases per day

```python
# Generate total number of covid cases per day in the United States
covid_cases_summed = refined_data.sum(axis = 0, skipna = True)
covid_cases_summed
```

```python
new_covid_dataframe = pd.DataFrame(columns=refined_data.columns)
new_covid_dataframe = new_covid_dataframe.append(covid_cases_summed, ignore_index=True)
```

## Number of cases per day after calculating difference from previous day with current day

```python
newcases_perday = new_covid_dataframe.diff(axis=1)
newcases_perday
```

```python
newcases_perday.drop(columns=['2020-01-22'], inplace=True)
newcases_perday
```

## Calculate the total number of weeks for the dataset

```python
import math
number_of_weeks = math.floor(len(covid_cases_summed)/7)
number_of_weeks
```

## Total number of cases every week

```python
# This function calculates the total number of cases and average cases occured every week based on the operation value recieved
def calculate_sum_avg(dataset,operation):
    weekly_data = []
    i=0
    counter=7
    x=0
    while (x<number_of_weeks):
        if(operation == "sum"):
            batch = dataset.iloc[:,i:counter].sum(axis=1).astype(int)
        elif (operation == "avg"):
            batch = round(dataset.iloc[:,i:counter].mean(axis=1)).astype(int)
        i = i+7
        counter = counter+7
        x = x+1
        weekly_data.append(batch)
    return weekly_data
```

```python
IS_SUM = "sum"
weekly_summed_data = calculate_sum_avg(newcases_perday,IS_SUM)
total_cases = pd.concat(weekly_summed_data, axis=1)
total_cases
```

##  Average number of cases every week

```python
IS_MEAN = "avg"
weekly_averaged_data = calculate_sum_avg(newcases_perday,IS_MEAN)
new_avg = pd.concat(weekly_averaged_data, axis=1)
new_avg
```

##  calculating (mean, median, mode) for number of covid cases of every week taken together.

```python
mean_all_week = round(new_avg.mean(axis=1)).astype(int)
mean_all_week
```

```python
median_all_week = round(new_avg.median(axis=1)).astype(int)
median_all_week
```

```python
mode_all_week = round(new_avg.mode(axis=1)).astype(int)
mode_all_week
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
import datetime
# Reading the world covid data which contains all country data
covid_world = pd.read_csv('../../../data/stage_2/owid-covid-data.csv')
covid_world
```

##### This dataset contain data of all countries, so selecting specific countries that we want to compare with (Indonesia, Brazil, Nigeria, Russia, Japan)

```python
covid_world = covid_world[covid_world["location"].isin(["Indonesia", "Brazil", "Nigeria", "Russia", "Japan"])]
covid_world                             
```

##### Selecting the specific data that we need from the whole dataset
##### some country values started from march, so to get data from same range of dates below date range values are chosen. Below dataframe shows each day cases and deaths for all 5 countries

```python
# Filtering the data
covid_world_data = covid_world[["location","date","new_cases","new_deaths","population"]]
covid_world_data = covid_world_data[(covid_world_data['date'] > '2020-03-10') & (covid_world_data['date'] <= '2021-08-16')]
covid_world_data.reset_index()
```

##### summing up the weekly data for each country. Below dataframe contains new cases and deaths value per week

```python
covid_world_data['date'] = pd.to_datetime(covid_world_data['date'])
covid_weekly = covid_world_data.groupby('location').resample('W-Mon', on='date').sum().reset_index().sort_values(by=['location','date'])
```

```python
# Normalizing it
covid_weekly['new_cases'] = covid_weekly['new_cases']/covid_weekly['population'] * 10000000
covid_weekly['new_deaths'] = covid_weekly['new_deaths']/covid_weekly['population'] * 10000000
covid_weekly
```

##### Calculating per day mean value for each country (for new cases and deaths as well)

```python
world_perday_mean = covid_world_data.groupby("location")["new_cases","new_deaths"].mean().round(decimals=0)
world_perday_mean
```

##### Calculating Mean of every week

```python
covid_week_mean = covid_world_data.groupby('location').resample('W-Mon', on='date').mean().reset_index().sort_values(by=['location','date'])
round(covid_week_mean)
```

#### Calculating mean value of all weeks taken together for each country. Inserting the US mean values to this dataframe for better comparison

```python
world_weekly_mean = covid_weekly.groupby("location")["new_cases","new_deaths"].mean().round(decimals=0).reset_index()
world_weekly_mean.loc[len(world_weekly_mean.index)] = ['UnitedStates', mean_all_week[0], mean_death_weekly[0]]
world_weekly_mean = world_weekly_mean.set_index('location')
world_weekly_mean
```

##### Calculating median value of all weeks for each country. Inserting the US median values to this dataframe for better comparison

```python
world_weekly_median = covid_weekly.groupby("location")["new_cases","new_deaths"].median().round(decimals=0).reset_index()
world_weekly_median.loc[len(world_weekly_median.index)] = ['UnitedStates', median_all_week[0], median_death_weekly[0]]
world_weekly_median= world_weekly_median.set_index('location')
world_weekly_median
```

```python
world_weekly_cases_mode = covid_weekly.mode()['new_cases'][0]
world_weekly_deaths_mode = covid_weekly.mode()['new_deaths'][0]
print("Mode of New Cases",world_weekly_cases_mode)
print("Mode of New Deaths",world_weekly_deaths_mode)
```

## Task 3 - Calculaing Trend for covid-19 Dataset

```python
read_usa_facts
```

## Reading enrichment Dataset - ACS Demographic and housing

```python
read_enrichment_dataset = pd.read_csv("../../../../data/stage_1/ACSDP1Y2019.DP05_data_with_overlays_2021-09-21T071853.csv")
```

```python
enrichment_dframe = pd.DataFrame(read_enrichment_dataset)
```

```python
enrichment_dframe
```

```python
enrichment_dframe.dtypes
```

```python
pd.isnull(enrichment_dframe)
```

```python
enrichment_dframe.isna().any()
```

```python
dropped_frame=enrichment_dframe.drop([0])
```

```python
dropped_frame
```

```python
dropped_frame['GEO_ID'] = dropped_frame['GEO_ID'].str.slice(start=9)
```

```python
dropped_frame = dropped_frame.rename(columns={"GEO_ID":"countyFIPS"})
```

```python
dropped_frame = dropped_frame.astype({'countyFIPS': 'int64'})
```

```python
dropped_frame.dtypes
```

```python
read_super_dataset = pd.read_csv("../../../../data/stage_1/Covid_merged_dataset.csv")
```

```python
read_super_dataset
```

```python
read_super_dataset.dtypes
```

```python
enrichment_covid_merge = pd.merge(read_super_dataset, dropped_frame, on ='countyFIPS')
```

```python
enrichment_covid_merge.to_csv('../../../../data/stage_1/enrichment_covid_merge_deepa.csv', index=False)
```

```python
import pandas as pd
confirmed_cases = pd.read_csv('../../../../data/stage_2/covid_confirmed_usafacts.csv')
death_cases = pd.read_csv('../../../../data/stage_2/covid_deaths_usafacts.csv')
population_dataset = pd.read_csv('../../../../data/stage_2/covid_county_population_usafacts.csv')
population_df = pd.DataFrame(population_dataset)
```

## Generate weekly statistics (mean, median, mode) for number of new cases and deaths across a specific state.

## Weekly statistics for everyday new cases in NC state

```python
nc_data = confirmed_cases[confirmed_cases["State"] == "NC"]
nc_data.head()
```

```python
#Resetting the index to below columns so that they don't get summed up
refined_data = nc_data.set_index(['countyFIPS','County Name','State','StateFIPS'])
refined_data.head()
```

## Number of cases per day

```python
# sum up the column values to get the total number of cases for the particular day and calculate difference to get new cases per day
covid_cases_summed = refined_data.sum()
new_covid_dataframe = pd.DataFrame(columns=refined_data.columns)
new_covid_dataframe = new_covid_dataframe.append(covid_cases_summed, ignore_index=True)
new_covid_dataframe
```

```python
new_covid_dataframe = new_covid_dataframe.diff(axis=1)
new_covid_dataframe = new_covid_dataframe.drop(['2020-01-22'], axis=1)
new_covid_dataframe
```

## Function to calculate weekly mean

```python
import math
number_of_weeks = math.floor(len(new_covid_dataframe.columns)/7)
def calculate_mean_median_mode(dataset,action):
    weekly_data = []
    i=0
    counter=7
    x=0
    while (x<number_of_weeks):
        if(action=="sum"):
            batch = dataset.iloc[:,i:counter].sum(axis=1).astype(int)
        i = i+7
        counter = counter+7
        x = x+1
        weekly_data.append(batch)
    return weekly_data
```

```python
IS_SUM = "sum"
sum_data = calculate_mean_median_mode(new_covid_dataframe,IS_SUM)
sum_cases_weekly = pd.concat(sum_data, axis=1)
sum_cases_weekly
```

## Average number of cases every week

```python
weekly_avg_data = sum_cases_weekly.mean(axis=1).astype(int)
weekly_avg_data
```

## Median for weekly cases

```python
median_week = sum_cases_weekly.median(axis=1).astype(int)
median_week
```

## Mode for weekly cases

```python
round(sum_cases_weekly.mode(axis=1)).astype(int)
```

## Covid 19 Death case analysis for NC state

```python
nc_data_death = death_cases[death_cases["State"] == "NC"]
nc_data_death.head()
```

```python
#Resetting the index to below columns so that they don't get summed up
refined_data_death = nc_data_death.set_index(['countyFIPS','County Name','State','StateFIPS'])
refined_data_death.head()
```

```python
# sum up the column values to get the total number of cases for the particular day and calculate difference to get new cases per day
covid_death_summed = refined_data_death.sum()
death_df = pd.DataFrame(columns=refined_data_death.columns)
death_df = death_df.append(covid_death_summed, ignore_index=True)
# Finding new cases everyday
death_df = death_df.diff(axis=1)
#dropping Nan value
death_df =death_df.drop(['2020-01-22'], axis=1)
death_df
```

```python
sum_data_death = calculate_mean_median_mode(death_df,IS_SUM)
death_weekly = pd.concat(sum_data_death, axis=1)
death_weekly
```

## Weekly death cases mean

## Weekly death cases median

```python
round(death_weekly.median(axis=1)).astype(int)[0]
```

## Weekly death cases mode

```python
round(death_weekly.mode(axis=1)).astype(int)[0][0]
```

## Daily new cases normalized by population

```python
state_wise_data = confirmed_cases[confirmed_cases['County Name'] != 'Statewide Unallocated']
state_wise_data = state_wise_data.set_index(['countyFIPS','County Name','State','StateFIPS'])
```

```python
# Take difference of every day to get the cases per day
everyday_new_cases = state_wise_data.diff(axis=1)
everyday_new_cases = everyday_new_cases.drop(columns = ['2020-01-22'])
```

```python
merged_cases_perday = pd.merge(everyday_new_cases, population_df, on='countyFIPS', how ='inner' )
merged_cases_perday = merged_cases_perday.set_index(['countyFIPS','State','County Name'])
merged_cases_perday.head()
```

```python
merged_cases_perday = round(((merged_cases_perday.div(merged_cases_perday['population'], axis=0))*100000).astype(int))
merged_cases_perday.head()
```

```python
merged_cases_perday = merged_cases_perday.drop(columns = ['population'])
#merged_cases_perday
```

```python
weekly_averaged_data = calculate_mean_median_mode(merged_cases_perday,"sum")
new_avg = pd.concat(weekly_averaged_data, axis=1)
new_avg.head()
```

```python
grouped_data = new_avg.groupby(['State']).sum()
grouped_data
```

```python
def calculate_mean_med_mode(df):
    mean_values = []
    median_values = []
    mode_values = []
    for index, row in df.iterrows():
        mean_values.append(round(row.mean()))
        median_values.append(round(row.median()))
        mode_values.append(round(row.mode()[0]))  
    df["mean"] = mean_values  
    df["median"] = median_values
    df["mode"] = mode_values
    return df        
```

```python
mean_med_mode = calculate_mean_med_mode(grouped_data)
mean_med_mode
```

## Identify counties within the previous state with high case and death rates. (normalize by population)

## Chosen state is NC for covid new case analysis

```python
nc_state_rates = confirmed_cases[confirmed_cases['State'] == "NC"]
nc_state_rates = nc_state_rates[nc_state_rates['County Name'] != 'Statewide Unallocated']
nc_state_rates = nc_state_rates.set_index(['countyFIPS','County Name','State','StateFIPS'])
nc_state_rates = nc_state_rates.diff(axis=1)
nc_state_rates = nc_state_rates.drop(columns = ['2020-01-22'])
nc_merged_pop = pd.merge(nc_state_rates, population_df, on='countyFIPS', how ='inner' )
nc_merged_pop = nc_merged_pop.set_index(['countyFIPS','State','County Name'])
nc_merged_pop = round(((nc_merged_pop.div(nc_merged_pop['population'], axis=0))*100000).astype(int))
nc_merged_pop = nc_merged_pop.drop(columns = ['population'])
weekly_avg_nc = calculate_mean_median_mode(nc_merged_pop,"sum")
#death_cases_avg
new_avg_nc = pd.concat(weekly_avg_nc, axis=1)
new_avg_nc['Avg_cases'] = round(new_avg_nc.mean(axis=1).astype(int))
sorted_cases = new_avg_nc.sort_values(by =['Avg_cases'], ascending=False)
sorted_cases.head(5)
```

```python
## Repeat the same procedure for Death rates as well
nc_state_death_county = death_cases[death_cases['State'] == "NC"]
nc_state_death_county = nc_state_death_county[nc_state_death_county['County Name'] != 'Statewide Unallocated']
nc_state_death_county = nc_state_death_county.set_index(['countyFIPS','County Name','State','StateFIPS'])
nc_state_death_county = nc_state_death_county.diff(axis=1)
nc_state_death_county = nc_state_death_county.drop(columns = ['2020-01-22'])
nc_merged_pop_county = pd.merge(nc_state_death_county, population_df, on='countyFIPS', how ='inner' )
nc_merged_pop_county = nc_merged_pop_county.set_index(['countyFIPS','State','County Name'])
nc_merged_pop_county = round(((nc_merged_pop_county.div(nc_merged_pop_county['population'], axis=0))*100000).astype(int))
nc_merged_pop_county = nc_merged_pop_county.drop(columns = ['population'])
weekly_avg_nc_county = calculate_mean_median_mode(nc_merged_pop_county,"sum")
#death_cases_avg
weekly_avg_nc_county = pd.concat(weekly_avg_nc_county, axis=1)
weekly_avg_nc_county['Avg_cases'] = round(weekly_avg_nc_county.mean(axis=1))
sorted_cases_county = weekly_avg_nc_county.sort_values(by =['Avg_cases'], ascending=False)
sorted_cases_county.head(5)
```

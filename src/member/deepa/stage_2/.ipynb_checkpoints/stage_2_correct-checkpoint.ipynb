```python
# Reading the covid-19 dataset
import pandas as pd
confirmed_cases = pd.read_csv('../../../../data/stage_2/covid_confirmed_usafacts.csv')
confirmed_cases
```

```python
state_wise_data = confirmed_cases[confirmed_cases["State"] == "NC"]
state_wise_data
```

```python
#Resetting the index to below columns so that they don't get summed up
refined_data = state_wise_data.set_index(['countyFIPS','County Name','State','StateFIPS'])
refined_data
```

```python
import math
number_of_weeks = math.floor(len(refined_data.columns)/7)
number_of_weeks
```

## Number of cases per day

```python
# sum up the column values to get the total number of cases for the particular day and calculate difference to get new cases per day
covid_cases_summed = refined_data.sum(axis = 0, skipna = True)
new_covid_dataframe = pd.DataFrame(columns=refined_data.columns)
new_covid_dataframe = new_covid_dataframe.append(covid_cases_summed, ignore_index=True)
new_covid_dataframe
```

```python
new_covid_dataframe = new_covid_dataframe.diff(axis=1)
```

```python
new_covid_dataframe =new_covid_dataframe.drop(['2020-01-22'], axis=1)
new_covid_dataframe
```

```python
import math
number_of_weeks = math.floor(len(new_covid_dataframe.columns)/7)
def calculate_mean_median_mode(dataset,action):
    weekly_data = []
    i=0
    counter=7
    x=0
    while (x<number_of_weeks):
        if(action == "mean"):
            batch = dataset.iloc[:,i:counter].mean(axis=1).astype(int)
        i = i+7
        counter = counter+7
        x = x+1
        weekly_data.append(batch)
    return weekly_data
    
```

```python
IS_AVG = "mean"
avg_data = calculate_mean_median_mode(new_covid_dataframe,IS_AVG)
avg_cases_weekly = pd.concat(avg_data, axis=1)
avg_cases_weekly
```

## Average number of cases every week

```python
round(avg_cases_weekly.mean(axis=1)).astype(int)
```

## Median of weekly cases

```python
round(avg_cases_weekly.median(axis=1)).astype(int)
```

## Mode for weekly cases

```python
round(avg_cases_weekly.mode(axis=1)).astype(int)
```

```python
# sum up the column values to get the total number of cases for the particular day and calculate difference to get new cases per day
covid_death_summed = refined_data_death.sum(axis = 0, skipna = True)
death_df = pd.DataFrame(columns=refined_data_death.columns)
death_df = death_df.append(covid_death_summed, ignore_index=True)
# Finding new cases everyday
death_df = death_df.diff(axis=1)
#dropping Nan value
death_df =death_df.drop(['2020-01-22'], axis=1)
death_df
```

```python
avg_data_death = calculate_mean_median_mode(death_df,IS_AVG)
death_weekly = pd.concat(avg_data_death, axis=1)
death_weekly
```

```python
round(death_weekly.median(axis=1)).astype(int)
```

## Compare the data against other States (Normalize by population)

```python
# To normalize the data by population, we need total population of the state
population_dataset = pd.read_csv('../../../../data/stage_2/covid_county_population_usafacts.csv')
population_df = pd.DataFrame(population_dataset)
population_df
```

```python
#reshaping the data to long format
reshaped_covid_cases = confirmed_cases.melt(id_vars=['countyFIPS', 'County Name', 'State', 'StateFIPS'],  var_name='Date', value_name='Covid_cases')
reshaped_covid_cases['Date'] = pd.to_datetime(reshaped_covid_cases['Date'])
reshaped_covid_cases
```

```python
reshaped_covid_cases = reshaped_covid_cases[reshaped_covid_cases['County Name'] != 'Statewide Unallocated'].reset_index(drop=True)
```

```python
reshaped_covid_cases
```

```python
pop_cases = pd.merge(reshaped_covid_cases,population_dataset, on ="countyFIPS",how="inner")
pop_cases = pop_cases.drop(columns=['County Name_y','State_y'])
pop_cases['Date'] = pd.to_datetime(pop_cases['Date'])
pop_cases.dtypes
```

```python
pop_cases = pop_cases.groupby(['State_x', 'County Name_x', 'Date', 'population'],as_index=False).agg({'Covid_cases': 'sum'})
pop_cases['Covid_cases'] = pop_cases['Covid_cases'].diff()
pop_cases =pop_cases.drop([0])
pop_cases
```

```python
cases_by_states = pop_cases.groupby(['State_x','Date']).agg({'Covid_cases':'sum','population':'sum'}).reset_index()
cases_by_states
```

```python
cases_by_states['Normalized_cases'] = round(cases_by_states['Covid_cases']/cases_by_states['population']*100000)
cases_by_states = cases_by_states.drop([0])
cases_by_states
```

```python
cases_by_states['Date'] = pd.to_datetime(cases_by_states['Date'])
cases_by_states.dtypes
```

```python
from datetime import datetime

cases_by_states['Date'] = cases_by_states['Date'].dt.strftime('%Y-%m-%d')
cases_by_states['Date'] = cases_by_states.Date.apply(lambda d: datetime.strptime(d, '%Y-%m-%d'))
cases_by_states = cases_by_states.groupby(['State_x', 'population', pd.Grouper(key='Date', freq='w')]).mean().reset_index().round(decimals=0)
cases_by_states
```
